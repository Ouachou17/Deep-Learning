```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical

# les DonnГ©es texte
text = [
    "je suis Г©tudiant en informatique",
    "j aime le machine learning",
    "le deep learning est intГ©ressant",
    "je travaille sur des rГ©seaux de neurones",
    "les rГ©seaux rГ©currents traitent les sГ©quences"
]


# on fait la Tokenisation
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text)
sequences = tokenizer.texts_to_sequences(text)
word_index = tokenizer.word_index

print("Dictionnaire :", word_index)

# la CrГ©ation des sГ©quences
input_sequences = []

for seq in sequences:
    for i in range(1, len(seq)):
        input_sequences.append(seq[:i+1])

# on fait le Padding
max_len = max([len(seq) for seq in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
input_sequences = np.array(input_sequences)

# SГ©paration de X et y
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# One-Hot Encoding
vocab_size = len(word_index) + 1
y = to_categorical(y, num_classes=vocab_size)

print("Shape X :", X.shape)
print("Shape y :", y.shape)


# on applique le ModГЁle RNN
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=max_len-1))
model.add(SimpleRNN(50))
model.add(Dense(vocab_size, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam')

model.summary()

# EntraГ®nement (Training)
model.fit(X, y, epochs=200, verbose=1)

# on fait la prГ©diction (Fonction de prГ©diction)
def predict_next_word(model, tokenizer, text, max_len):
    sequence = tokenizer.texts_to_sequences([text])[0]
    sequence = pad_sequences([sequence], maxlen=max_len-1, padding='pre')

    predicted = model.predict(sequence, verbose=0)
    predicted_index = np.argmax(predicted)

    for word, index in tokenizer.word_index.items():
        if index == predicted_index:
            return word
    return ""


# la partie de Testing
# ================================
print("\nTests de prГ©diction :")
print("je suis ->", predict_next_word(model, tokenizer, "je suis", max_len))
print("le deep ->", predict_next_word(model, tokenizer, "le deep", max_len))
print("les rГ©seaux ->", predict_next_word(model, tokenizer, "les rГ©seaux", max_len))
```
